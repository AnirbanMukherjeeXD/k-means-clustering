#K-Means Clustering on Iris Dataset
###Using TSNE for dimension reduction

```
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

**Load Dataset**

```
dataset=load_iris()
X=dataset.data
```

##2 Dimensions

```
X= TSNE(n_components=2).fit_transform(X)
model = KMeans(n_clusters=3, random_state=0)
Y=model.fit_predict(X)
```

![alt text](https://github.com/AnirbanMukherjeeXD/k-means-clustering/blob/master/Clusters-3D.gif)

```
y=dataset.target
plt.scatter(X[y==0,0],X[y==0,1],marker='o',alpha=0.7)
plt.scatter(X[y==1,0],X[y==1,1],marker='o',alpha=0.7)
plt.scatter(X[y==2,0],X[y==2,1],marker='o',alpha=0.7)
plt.scatter(X[Y==0,0],X[Y==0,1],c='red',marker='x')
plt.scatter(X[Y==1,0],X[Y==1,1],c='blue',marker='x')
plt.scatter(X[Y==2,0],X[Y==2,1],c='green',marker='x')
```

```
plt.scatter(X[Y==0,0],X[Y==0,1],c='red')
plt.scatter(X[Y==1,0],X[Y==1,1],c='blue')
plt.scatter(X[Y==2,0],X[Y==2,1],c='green')
```

##3 Dimensions

```
X= TSNE(n_components=3).fit_transform(X)
model = KMeans(n_clusters=3, random_state=42)
Y=model.fit_predict(X)

fig = plt.figure()
ax = fig.gca(projection='3d')
c=['red','blue','green']
for i in range(3):
  ax.scatter(X[Y==i,0],X[Y==i,1],X[Y==i,2],alpha=1,color=c[i])
fig.show()
```
